```python
import math

def calculate_triangle_area(a, b, c):
    # Calculate the semi-perimeter
    s = (a + b + c) / 2

    # Calculate the area using Heron's formula
    area = math.sqrt(s * (s - a) * (s - b) * (s - c))
    return area

# Example usage:
side1 = 3
side2 = 4
side3 = 5

area_result = calculate_triangle_area(side1, side2, side3)
print(f"The sides of the triangle are: {side1}, {side2}, {side3}")
print(f"The area of the triangle is: {area_result}")
```

```python
def classify_character(char_input):
    if char_input.isdigit():
        return f"The character '{char_input}' is a digit."
    elif char_input.islower():
        return f"The character '{char_input}' is a lowercase letter."
    elif char_input.isupper():
        return f"The character '{char_input}' is an uppercase letter."
    else:
        return f"The character '{char_input}' is none of the above (not a digit, lowercase, or uppercase letter)."

print("Function 'classify_character' defined.")
```

```python
test_characters = ['7', 'k', 'P', '@', '$', ' ', 'a', 'Z', '9', '!']

for char in test_characters:
    print(classify_character(char))
```

```python
raw_input = input('Enter a series of integers separated by spaces or commas: ')

# Replace commas with spaces to handle both delimiters, then split and convert
user_numbers = [int(x) for x in raw_input.replace(',', ' ').split()]

print(f'Captured list of integers: {user_numbers}')
```

```python
def calculate_factorial(n):
    if n < 0:
        return None
    if n == 0:
        return 1
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

# Calculate factorials for the numbers in user_numbers
user_factorials = [calculate_factorial(num) for num in user_numbers]

print(f"User numbers: {user_numbers}")
print(f"Calculated factorials: {user_factorials}")
```

```python
for num, fact in zip(user_numbers, user_factorials):
    print(f'The factorial of {num} is {fact}')

print('\nFull list of factorials generated for user input:')
print(user_factorials)
```
